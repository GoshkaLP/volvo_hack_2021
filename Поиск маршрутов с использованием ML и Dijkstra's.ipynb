### Импорт билиотек

```python
import base64

import haversine as hs
from haversine import Unit

from dijkstar import Graph, find_path

import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

import numpy as np
import pandas as pd

from sklearn.cluster import KMeans

import folium
from folium import IFrame
from folium.plugins import MarkerCluster
```

### Объекты ЮНЕСКО

```python
# Данные были скачаны с официального сайта ЮНЕСКО
data = pd.read_excel('data/unesco_2021.xls')

data = data[
    data['states_name_en'] == 'Russian Federation'
].reset_index(drop=True)

data['short_description_en'] = data['short_description_en'].str\
                               .replace('<p>', '').replace('</p>', '')

data.head(3).T
```

### Интерактивная карта объектов ЮНЕСКО

#### Логотип ЮНЕСКО

```python
unesco_logo = base64.b64encode(open('img/unesco_logo.png', 'rb').read())
unesco_logo = unesco_logo.decode('UTF-8')
```

#### Всплывающие окна на карте

```python
unesco_popups = []

for index, row in data.iterrows():
    name = row['name_en']
    description = row['short_description_en']
    
    html = f"""
        <b> NAME: </b> {name} <br>  <p>
        <b> DESCRIPTION: </b> {description} <br>
        <center> <img src="data:image/png;base64,{unesco_logo}"> </canter>
    """
    
    iframe = IFrame(html, width = 500, height = 300)
    
    popup = folium.Popup(iframe)
    unesco_popups.append(popup)
```

#### Координаты объектов

```python
lats = data['latitude'].to_list()
lons = data['longitude'].to_list()

locations = list(zip(lats, lons))
```

#### Добавление объектов на карту

Все объекты, как и сама карта, являются интерактивными. 
Если увеличивать/отдалять карту объекты буду разъединяться/группироваться в классы в зависимости от того как они рассположены относительно друг друга.
При нажатии на объект выводится базовая информация по нему, полученная из открытых источников:
* Название
* Описание
* Изображение

```python
m = folium.Map(
    location = [np.mean(lats), np.mean(lons)], 
    tiles = "Cartodb Positron", zoom_start = 3
)

marker_cluster = MarkerCluster(
    locations = locations,
    popups = unesco_popups,
    name = "Объекты ЮНЕСКО",
    overlay = True,
    control = True,
)

marker_cluster.add_to(m)
folium.LayerControl().add_to(m)

m
```

### Кластеризация объектов

#### Определение оптимального количества классов

```python
sum_of_squared_distances = []
K = range(1, 10)

for num_clusters in K :
    kmeans = KMeans(n_clusters = num_clusters)
    kmeans.fit(locations)
    sum_of_squared_distances.append(kmeans.inertia_)
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
plt.plot(K, sum_of_squared_distances, 'bx-')

plt.xlabel('Values of K') 
plt.ylabel('Sum of squared distances/Inertia') 
plt.title('Elbow Method For Optimal k')

plt.show()
```

#### Обучение модели

```python
kmeans = KMeans(n_clusters = 3, random_state = 2021)
kmeans.fit(locations)
```

```python
data['kmeans_class'] = kmeans.predict(locations)

data['kmeans_class'].value_counts().to_frame('count')\
.reset_index().sort_values('index')
```

#### Визуализация результатов

```python
colors = mcolors.TABLEAU_COLORS
colors = list(colors.values())
print(*colors)
```

```python
m = folium.Map(
    location = [56.326722, 44.006479], 
    tiles = "Cartodb Positron", 
    zoom_start = 6
)


for i, row in data.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    kmeans_class = row['kmeans_class']
    
    folium.Circle(
        [latitude, longitude],
        color = colors[7 - kmeans_class],
        fill_color = colors[7 - kmeans_class],
        fill_opacity = 1,
        radius=5000,
    ).add_to(m)

m
```

### Построение кратчайшего маршрута

#### Координаты маршрута

```python
start_point = [55.679138, 37.263658]
stop_point = [56.091329, 49.876984]
```

#### Дополнительные элементы на карте

```python
popup_start = "Ура! Поехали."
popup_stop = "Ура! Доехали."

tooltip_start = "Начало маршрута"
tooltip_stop = "Конец маршрута"
```

#### Добавление объектов на карту

```python
m = folium.Map(
    location = [
        np.mean([start_point[0], stop_point[0]]), 
        np.mean([start_point[1], stop_point[1]]),
    ], 
    tiles = "Cartodb Positron", zoom_start = 6
)

marker_cluster = MarkerCluster(
    locations = locations,
    popups = unesco_popups,
    name = "Объекты ЮНЕСКО",
    overlay = True,
    control = True,
).add_to(m)

folium.Marker(
    start_point, 
    popup = popup_start, 
    tooltip = tooltip_start, 
    icon = folium.Icon(color = "red", icon = "info-sign")
).add_to(m)


folium.Marker(
    stop_point, 
    popup = popup_stop, 
    tooltip = tooltip_stop, 
    icon=folium.Icon(color = "red", icon = "info-sign")
).add_to(m)

folium.PolyLine([start_point, stop_point], color = 'red').add_to(m)
folium.LayerControl().add_to(m)

m
```

### Построение интересных маршрутов

#### Расстояние между объектами

```python
def get_distance(row):
    loc1 = (row['lats_start'], row['lons_start'])
    loc2 = (row['lats_stop'], row['lons_stop'])
    
    return round(hs.haversine(loc1,loc2,unit = Unit.METERS))
```

```python
# Вычисляем расстояния между всеми объектами.
# При расчете используется алгоритм: Haversine.

distance_between_object = pd.DataFrame({
    'lats': [start_point[0], stop_point[0]] + lats,
    'lons': [start_point[1], stop_point[1]] + lons,
    'name': ['Name: Star Point', 'Name: End Points'] + [name for name in data['name_en']],
    'kmeans_class': [None, None] + data['kmeans_class'].to_list()
}).reset_index()

distance_between_object['merge'] = 1
distance_between_object = distance_between_object.merge(
    distance_between_object,
    on = 'merge',
    suffixes=('_start', '_stop')
)

distance_between_object['distance'] = distance_between_object.apply(get_distance, axis = 1)
distance_between_object[distance_between_object['index_start'] == 2]
```

#### Фактор кластеризации

```python
# Если объекты расположены рядом относительно друг друга, то
# искусственно уменьшаем расстояние между ними. Это дает возможность
# алгоритму рассмотреть как можно больше интересных маршрутов.

distance_upd = []

for i, row in distance_between_object.iterrows():
    if (row['index_start'] != row['index_stop']) & (row['kmeans_class_start'] == row['kmeans_class_stop']):
            distance_upd.append(row['distance'] * .9)
    else:
        distance_upd.append(row['distance'])

        
distance_between_object['distance_upd'] = distance_upd
```

#### Создание взвешенного графа

```python
graph = Graph()

for i, row in distance_between_object.iterrows():
    graph.add_edge(row['index_start'], row['index_stop'], row['distance_upd'])
```

#### Поиск маршрутов

```python
routes = [find_path(graph, 0, 1)]
```

#### Создание маршрута

```python
uniqie_object = distance_between_object[
    distance_between_object.columns[:4]
].drop_duplicates().reset_index(drop = True)

route_coordinates = []

for i in routes[0].nodes:
    point = [
        uniqie_object[uniqie_object['index_start'] == i]['lats_start'].iloc[0],
        uniqie_object[uniqie_object['index_start'] == i]['lons_start'].iloc[0]
    ]
    
    route_coordinates.append(point)
```
